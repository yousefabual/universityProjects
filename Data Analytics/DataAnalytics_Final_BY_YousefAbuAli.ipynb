## **Applying methods to remove outliers:**

```python
HIGH=['VOLUME','TRADE_QTY','NO_OF_TRADES','HIGH','BEST_ASK_PRICE','BEST_ASK_QTY','BEST_BID_PRICE','BEST_BID_QTY']
LOW=['VOLUME','TRADE_QTY','NO_OF_TRADES','LOW','BEST_ASK_PRICE','BEST_ASK_QTY','BEST_BID_PRICE','BEST_BID_QTY']
```

### **interquartile range method:**

**For the HIGH Dataset:**

```python
# the following lines calculates the first and thire quartile range and findes the interquartile range
# this is used to find outliers
Q1 = StocksDataSetHigh[HIGH].quantile(0.25)
Q3 = StocksDataSetHigh[HIGH].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
```

```python
# this line indentify outliers by making sure that we remove outliers less than lower and higher than upper
outliers = ((StocksDataSetHigh[HIGH] < lower) | (StocksDataSetHigh[HIGH] > upper)).any(axis=1)
```

```python
# this line removes the outliers from the dataset and store it in df_cleaned
df_cleaned1 = StocksDataSetHigh[HIGH] [~outliers]
```

**For the LOW Dataset:**

```python
Q1 = StocksDataSetLow[LOW].quantile(0.25)
Q3 = StocksDataSetLow[LOW].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
```

```python
outliers = ((StocksDataSetLow[LOW] < lower) | (StocksDataSetLow[LOW] > upper)).any(axis=1)
```

```python
df_cleaned2 = StocksDataSetLow[LOW] [~outliers]
```

**Searching if their is still outliers:**

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 6))

# First boxplot
sns.boxplot(y='LOW', data=df_cleaned2, ax=axes[0])
axes[0].set_title('Boxplot - LOW dataset')

# Second boxplot
sns.boxplot(y='HIGH', data=df_cleaned1, ax=axes[1])
axes[1].set_title('Boxplot - HIGH dataset')

# set the space between plots
plt.subplots_adjust(wspace=0.4)

# Show the plot
plt.show()
```

### **First method: Remove features with low variance**

```python
# importing the necessary libraries that helps in feature selection and modling
from sklearn.feature_selection import VarianceThreshold
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# using the train_test_split() finction I will split the data into training and testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# I will calculate the variance for each feature in the dataset 'X':
# 1.helps to understand the spread of data
# 2.features with low variance are considered as noise
variances = X.var()
```

```python
print(variances)
```

```python
# setting an instance of the VarianceThreshold class to help us remove the features with low variance
# any feature with a variance equal to 45 or less will be removed
low_variance = VarianceThreshold(threshold = 45)
```

```python
# I will use the fit() function to fit the low_variance instance to the 'training' data.
# fit() function here helps to fit the low_variance object to the training data X_train, which will help in
# removing features with low variance
low_variance.fit(X_train)
```

```python
# now I will use the get_support() method, this method will give a sequence of true and false values:
# True indicates that a certain feature was selected, while false indicates that a feature was not.
# and this process depends on the low_variance results.
# The true, false values 'Boolean mask' will be used to retrive the columns that passed the test of variance.
selected_features = X.columns[low_variance.get_support()]
```

```python
# Now the selected fetures and there variance will be printed using a for loop that goes in the 'selected_features'
print('the selected features and there variance are as follows:')

for feature in selected_features:
  print(feature,': ',variances[feature])
```

```python
X_train_selected = low_variance.transform(X_train)
X_test_selected = low_variance.transform(X_test)
```

```python
# transform the entire dataset to keep only the selected features
X_selected = low_variance.transform(X)
```


**Linear Regression**


```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import matplotlib.pyplot as plt

# initialize a list to store the results
metr_results = []

for i in range(30):
    # split the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.2, random_state=i)

    # Next I will train a linear regression model using the feature that were selected:
    reg = LinearRegression().fit(X_train, y_train)
    # Next is to predict the variable y using a test set
    # predict is a method that helps in predicting the target variable
    y_pred = reg.predict(X_test)

   # y_pred 'predicted target values 'will be used to compare it with the true target value y_test, in order to calculate
  # the regression metrics
    metrics = {'MAE': mean_absolute_error(y_test, y_pred),
               'MSE': mean_squared_error(y_test, y_pred),
               'RMSE': mean_squared_error(y_test, y_pred, squared=False),
               'R2': r2_score(y_test, y_pred)}
    metr_results.append(metrics) # adding the result of the metrics after each loop to a list

# Storing the result in pandas dataFrame to use it easily for visualization
metricsData=pd.DataFrame(metr_results)

# using a box plot to plot the metrics

for column in metricsData.columns:
    plt.figure(figsize=(5,5))
    plt.boxplot(metricsData[column], vert=True)
    plt.title(column)
    plt.show()

# creating a dictionary to store the metrics to use them later on in the over all visualization:
Results['High LR using low variance']={'MAE': round(metricsData['MAE'].mean(),3),
                             'MSE': round(metricsData['MSE'].mean(),3),
                             'RMSE': round(metricsData['RMSE'].mean(),3),
                             'R2': round(metricsData['R2'].mean(),3)}
```

```python
# printing the result of the metrics
print(Results['High LR using low variance'])
```
