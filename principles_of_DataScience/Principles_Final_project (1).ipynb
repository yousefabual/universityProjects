### **Data Cleaning**

```python
# I_Vis represents how much space is dedicated to displaying a specific product within a store
# therefore it shouldn't contain zeros
#food['I_Vis']=food['I_Vis'].replace(0,'unkown')
food['I_Vis'] = food['I_Vis'].fillna(food['I_Vis'].mean())
```

```python
# filling the missing values in the wight column with the mean
food['I_W'] = food['I_W'].fillna(food['I_W'].mean())
```

```python
# filling the missing values in the size column with the mode
food['O_Size'] = food['O_Size'].fillna(food['O_Size'].mode()[0])
```

```python
# dropping the unnecessary features
food.drop('I_Id', axis=1, inplace=True)
food.drop('O_Id', axis=1, inplace=True)

food.drop('I_MRP_JD', axis=1, inplace=True)# because the final sales are in dollars
```

```python
food
```

### **Data Integration**

```python
food.I_Fat_C.unique()
```

```python
#make all the values in 'I_Fat_C' (Low Fat,Regular)
food['I_Fat_C'].replace({'low fat':'Low Fat','LF':'Low Fat','reg':'Regular'},inplace = True)
```

### **Data Transformation**

***using one hot encoder***

```python
# using one hot encoder for nominal data
ho_en_food1 = pd.get_dummies(food[['I_Fat_C']])
ho_en_food2 = pd.get_dummies(food[['I_category']])
#ho_en_food3 = pd.get_dummies(food[['O_T']]) dependent variable
ho_en_food4 = pd.get_dummies(food[['I_Recalled']])
```

```python
# adding them to the original dataset

food = food.drop(['I_Fat_C'],axis = 1 )
food = pd.concat([food,ho_en_food1],axis = 1)

food = food.drop(['I_category'],axis = 1 )
food = pd.concat([food,ho_en_food2],axis = 1)

food = food.drop(['I_Recalled'],axis = 1 )
food = pd.concat([food,ho_en_food4],axis = 1)
```

***Using label encoder***

```python
# using label encoder for categorical variable
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
food['O_Size'] = LE.fit_transform(food['O_Size'])
food['O_Loc_T'] = LE.fit_transform(food['O_Loc_T'])
```

### **Data Normalization**

**Data Normalization using Min-Max for the KNN algorithms**

```python
foodKNN['I_Vis']=(foodKNN['I_Vis']-foodKNN['I_Vis'].min())/(foodKNN['I_Vis'].max()-foodKNN['I_Vis'].min())
foodKNN['I_MRP_US']=(foodKNN['I_MRP_US']-foodKNN['I_MRP_US'].min())/(foodKNN['I_MRP_US'].max()-foodKNN['I_MRP_US'].min())
foodKNN['I_O_Sales']=(foodKNN['I_O_Sales']-foodKNN['I_O_Sales'].min())/(foodKNN['I_O_Sales'].max()-foodKNN['I_O_Sales'].min())  # ??
```
